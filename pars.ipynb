```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from string import punctuation
import pandas as pd
from bs4 import BeautifulSoup
```

```python
from pymorphy2 import MorphAnalyzer
pymorphy2_analyzer = MorphAnalyzer()

nltk.download('punkt');
nltk.download('stopwords');
```

```python
def tok_list(list_doc): #метод для токенизации заданного списка текстов
    llist_tok_words = []
    for doc in list_doc:
        llist_tok_words.append(word_tokenize(doc))
    return llist_tok_words

def preview_list(list_prevw): #метод для просмотра списков 
    for lst in list_prevw:
        print(lst)

def find_punk(list_doc): 
    llst_punkt = []
    for d in list_doc:
        for i in d:
            if ((len(i) == 1) and not(i in set('0123456789абвгдежзийклмнопрстуфхцчшщъыьэюяБВГДЕЖЗИЙКЛМНОПРСТУФХЦЧШЩЪЫЬЭЮЯ'))): #подумать над элегантным решением
                llst_punkt.append(i)
    return llst_punkt

def clear_punk(llist_tok_words, punk):
    str_punk = '!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~'
    llst_clear_punk = []
    if punk != '':
        str_punk = str_punk + punk
    for lst in llist_tok_words:
        llst_clear_punk.append(list(set(lst).difference(str_punk)))
    return llst_clear_punk

def lemmatizator(llist_tok_words):
    llst_lem = []
    lst_tmp = [] 
    for lst in llist_tok_words:
        for word in lst:
            lst_tmp.append(pymorphy2_analyzer.parse(word)[0].normal_form)
        llst_lem.append(lst_tmp)
        lst_tmp = []
    return llst_lem

def gen_uniform_dict(llist_tok_words):
    gen_set = set()
    for lst in llist_tok_words:
        gen_set = gen_set.union(set(lst))
    return gen_set    
```
