```python
path ="/home/syava/my_project_dir/data_book" 
list_dir = []
for root, dirs, files in os.walk(path): 
    for file in files: 
        if(file.endswith(".txt")): 
            res = os.path.join(root,file)
            res = res.split('/')
            list_dir.append(res[5])
```

```python
list_name = []
classes = []
book = []
for items in list_dir:
    lst = re.findall(r'_\w+_', items)
    for itt in lst:
        if len(itt)<7:
            classes.append(itt.replace('_', ''))
            book.append(items)
        else:
            list_name.append('_NAN_')
```

```python
list_index = []
for items in classes:
    if items[0] == 'M':
        list_index.append('M')
    elif items[0] == 'C':
        list_index.append('C')
    elif items[0] == 'G':
        list_index.append('G')
    else:
        list_index.append(np.NaN)
    
dic_b = {'Название': book, 'Класс': list_index}
```
