```python
def term_ex(term):
    term_extractor = TermExtractor()
    text = list_res[0][1]
    list_term = []
    list_count = []
    for term in term_extractor(term):
        list_term.append(term.normalized)
        list_count.append(term.count)
    dic_term = {'Словосочетания': list_term, 'Кол-во повторений': list_count}
    df = pd.DataFrame.from_dict(dic_term, orient='index')
    df_term = df.T
    df_term
    return(df_term)
```
