# функция для нахождения ключевых слов

```python
def term_ex(term):
    term_extractor = TermExtractor()
    list_term = []
    list_count = []
    for term in term_extractor(term, nested=True):
        list_term.append(term.normalized)
        list_count.append(term.count)
    dic_term = {'Словосочетания': set(list_term), 'Кол-во повторений':  set(list_count)}
    df = pd.DataFrame.from_dict(dic_term, orient='index')
    df_term = df.T
    df_term.sort_values(by=['Кол-во повторений'])
    df_term
    return(dic_term)
```

# визуализация в Pandas для словаря

```python
def vis_pd(dic_dic):
    df = pd.DataFrame.from_dict(dic_dic, orient='index')
    df_term = df.T
    df_term.sort_values(by=['Кол-во повторений'])
    return(df_term)
```
